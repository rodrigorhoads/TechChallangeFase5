```python
# 1. CONECTAR AO GOOGLE DRIVE
from google.colab import drive
drive.mount('/content/drive')

# --- CONFIGURAÇÃO ---
# Troque 'Hackathon_Vision' pelo nome EXATO da pasta que você criou no Drive
PASTA_DO_PROJETO = '/content/drive/MyDrive/fase5/diagramas/'

# 2. INSTALAR YOLO
!pip install ultralytics -q
from ultralytics import YOLO
import yaml
import os

# 3. VERIFICAÇÃO DE SEGURANÇA
# Vamos checar se o Colab está achando sua pasta
if not os.path.exists(PASTA_DO_PROJETO):
    print(f"ERRO: A pasta {PASTA_DO_PROJETO} não foi encontrada!")
    print("Verifique se o nome está correto e se você montou o Drive.")
else:
    print(f"Pasta encontrada! Caminho: {PASTA_DO_PROJETO}")

    # 4. REESCREVER O ARQUIVO DATA.YAML
    # O YOLO precisa de caminhos absolutos. Vamos criar um novo yaml temporário
    # apontando para o seu Drive corretamente.

    yaml_path_original = os.path.join(PASTA_DO_PROJETO, 'data.yaml')

    # Lendo o original para pegar os nomes das classes
    try:
        with open(yaml_path_original, 'r') as f:
            data_original = yaml.safe_load(f)
            names_classes = data_original.get('names', ['DATA_STORE', 'SERVER', 'USER', 'SECURITY_GATE'])
            nc_count = data_original.get('nc', 4)
    except:
        # Se der erro ao ler, usamos o padrão do Hackathon
        names_classes = ['DATA_STORE', 'SERVER', 'USER', 'SECURITY_GATE']
        nc_count = 4

    # Criando o novo conteúdo com o caminho certo do Drive
    novo_yaml_content = {
        'train': os.path.join(PASTA_DO_PROJETO, 'train/images'),
        'val': os.path.join(PASTA_DO_PROJETO, 'valid/images'),
        'nc': nc_count,
        'names': names_classes
    }

    # Salvamos um arquivo temporário no Colab para usar no treino
    yaml_temp_path = '/content/drive/MyDrive/fase5/config_drive.yaml'
    with open(yaml_temp_path, 'w') as f:
        yaml.dump(novo_yaml_content, f)

    print("Configuração gerada: apontando para as imagens no Drive.")

    # 5. INICIAR O TREINAMENTO
    print("Iniciando treinamento... (Isso pode demorar um pouco mais lendo do Drive)")
    model = YOLO('yolov8n.pt')

    results = model.train(
        data=yaml_temp_path, # Usamos o yaml que acabamos de criar
        epochs=50,
        imgsz=640,
        name='modelo_hackathon_drive',
        project='/content/drive/MyDrive/fase5/' # Salva o resultado direto no seu Drive
    )

    print(f"Sucesso! Seu modelo 'best.pt' estará salvo na pasta 'runs' dentro do seu projeto no Drive.")
```
